```python
def sum_even_numbers_explicit(numbers):
    total=0
for num in numbers:
    if num % 2 == 0:
        total += num
return total
```

```python
#2
def reverse_string(s):
    """
    Returns the reverse of the input string.
    
    Args:
    s (str): The input string to reverse.
    
    Returns:
    str: The reversed string.
    """
    return s[::-1]
```

```python
#3
def square_list(num):
    """
    Returns a new list containing the squares of each number in the input list.
    
    Args:
    num (list): List of integers
    
    Returns:
    list: List of squared integers
    """
    return [n ** 2 for n in num]
```

```python
#5
def fibonacci(n):
    
    l = [0,1] 
    for i in range(2,n):
        l.append(l[-1]+l[-2])
    return l
if __name__ == "__main__":
    n = 10
    fibo = fibonacci(n) 
    print("Fibonacci Series: ",*fibo)
```

```python
#9
def celsius_to_fahrenheit(celsius):
    """Convert Celsius to Fahrenheit."""
    return (celsius * 9/5) + 32
def main():
    try:
       
        user_input = input("Enter temperatures in Celsius (comma-separated): ").strip()
        celsius_list = []
        for temp in user_input.split(","):
            temp = temp.strip()
            if temp:  # Avoid empty strings
                try:
                    celsius_list.append(float(temp))
                except ValueError:
                    print(f"Invalid temperature value: '{temp}'. Skipping it.")
        
        if not celsius_list:
            print("No valid temperatures to convert.")
            return
        
        # Use map() to convert each Celsius value to Fahrenheit
        fahrenheit_list = list(map(celsius_to_fahrenheit, celsius_list))
        print("\nCelsius to Fahrenheit Conversion:")
        for c, f in zip(celsius_list, fahrenheit_list):
            print(f"{c}°C  ->  {f:.2f}°F")
    
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    main()
```

```python
#10
def remove_vowels(input_str):
    vowels = "aeiouAEIOU"
    result = ''.join(filter(lambda ch: ch not in vowels, input_str))
    return result

if __name__ == "__main__":
    try:
        user_input = input("Enter a string: ").strip()
        if not user_input:
            print("Empty string entered. Nothing to process.")
        else:
            no_vowels = remove_vowels(user_input)
            print("String without vowels:", no_vowels)
```

```python
#11
# Sample data: [Order Number, Price per Item, Quantity]
""" map() applies the lambda function to each order.
The lambda calculates price × quantity.
If the total is less than €100, it adds €10.
Returns a list of tuples (order_number, final_amount)."""
orders = [
    [34587, 40.95, 4],
    [98762, 56.80, 5],
    [77226, 32.95, 3],
    [88112, 24.99, 3]
]
result = list(map(lambda order: (
    order[0], 
    order[1] * order[2] if order[1] * order[2] >= 100 else order[1] * order[2] + 10
), orders))
print(result)
```
